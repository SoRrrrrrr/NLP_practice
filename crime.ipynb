```python
# 서울시 경찰청의 소속 구 @ https://goo.gl/MQSqXX
police_to_gu = {'서대문서': '서대문구', '수서서': '강남구', '강서서': '강서구', '서초서': '서초구',
                '서부서': '은평구', '중부서': '중구', '종로서': '종로구', '남대문서': '중구',
                '혜화서': '종로구', '용산서': '용산구', '성북서': '성북구', '동대문서': '동대문구',
                '마포서': '마포구', '영등포서': '영등포구', '성동서': '성동구', '동작서': '동작구',
                '광진서': '광진구', '강북서': '강북구', '금천서': '금천구', '중랑서': '중랑구',
                '강남서': '강남구', '관악서': '관악구', '강동서': '강동구', '종암서': '성북구', 
                '구로서': '구로구', '양천서': '양천구', '송파서': '송파구', '노원서': '노원구', 
                '방배서': '서초구', '은평서': '은평구', '도봉서': '도봉구'}
```

##### 경찰서를 구별로 정리

```python
#lambda => column 내 데이터마다 적용할 함수
df['구별']=df['관서명'].apply(lambda x: police_to_gu.get(x,'구 없음'))
df.head()
```

```python
#pivot_table ; index를 새로 지정하여 new df를 생성함 , aggfunc ; aggregation function 집계 함수
gu_df=pd.pivot_table(df,index='구별',aggfunc=np.sum)
gu_df
```

```python
gu_df=gu_df.drop(['구 없음'])
gu_df
```

##### 범죄별로 검거율 계산
검거율 = 검거 건수/발생 건수*100

```python
gu_df['강간 검거율']=gu_df['강간(검거)']/gu_df['강간(발생)']*100
gu_df['강도 검거율']=gu_df['강도(검거)']/gu_df['강도(발생)']*100
gu_df['살인 검거율']=gu_df['살인(검거)']/gu_df['살인(발생)']*100
gu_df['절도 검거율']=gu_df['절도(검거)']/gu_df['절도(발생)']*100
gu_df['폭력 검거율']=gu_df['폭력(검거)']/gu_df['폭력(발생)']*100
gu_df['검거율']=gu_df['소계(검거)']/gu_df['소계(발생)']*100
gu_df.head()
```

```python
columns=['강간 검거율','강도 검거율','살인 검거율','절도 검거율','폭력 검거율']
gu_df_rate=gu_df[columns]

#iterrows 각각의 행을 순서대로 꺼내주는 함수
for row_name, row in gu_df_rate.iterrows():
    for column in columns:
        if row[column] > 100:
            #.at[row, column] ; 해당 자리의 value값을 알 수 있음
            gu_df.at[row_name,column] =100

gu_df.head()
```

```python
#Masking.ver ; "체"를 활용하는 기법
gu_df[gu_df[['강간 검거율','강도 검거율', '살인 검거율','절도 검거율','폭력 검거율']]>100]=100
gu_df.head()
```

```python
#살인 발생 건수가 7건이 넘는 지역구 추려내기
gu_df[gu_df['살인(발생)']>7]
```

```python
gu_df[(gu_df['살인(발생)']>7) & (gu_df['폭력(발생)']>2000)]
```

```python
gu_df[(gu_df['살인(발생)']>7) | (gu_df['폭력(발생)']>2000)]
```

```python
gu_df[~(gu_df['살인(발생)']>5)]
```

```python
#'살인 검거율' 열의 결측치(NaN, Not a Number)를 다른 값으로 채워주기
gu_df['살인 검거율'].fillna(1000)
```

```python
# 새롭게 이름을 지어줄 떄 => rename
gu_df.rename(columns={'강간(발생)':'강간',
                      '강도(발생)':'강도',
                      '살인(발생)':'살인',
                      '절도(발생)':'절도',
                      '폭력(발생)':'폭력'
                      },inplace=True)
gu_df.head()
```

인구 데이터 merge 하기
A,B(type = pd.DataFrame) merge

1. A.join(B)
- 위아래로 섞여 있어도 제자리로 잘 합쳐줌
- A, V DataFrame의 index column이 동일해야 함.

2. pd.merge(A,B,left_on='구별', right_on='구이름', how='inner')
- 위아래로 섞여있어도 제자리로 잘 합쳐줌
- 기준이 되는 열이 서로 다른 이름을 가지고 있어도 ㄱㅊ.

3. pd.concat([A,B],axis=1)
- 위아래로 섞여있으면 그대로 합쳐줌.
- 행방향 or 열방향으로 적용 가능

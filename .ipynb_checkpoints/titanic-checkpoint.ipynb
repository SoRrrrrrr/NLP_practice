# 캐글 타이타닉 : 역사상 최대 해난사고 데이터 
탑승객의 생존여부 예측 모델을 구하는 것

dataset 확인 : 대부분의 캐글 데이터들은 잘 정제되어 있음. but, 가끔 null data가 존재함. 이를 확인하고, 향후 수정

탐색적 데이터 분석 : 여러 feature 들을 개별적으로 분석하고, feature들 간의 상관관계를 확인. 여러 시각화 툴을 사용하여 insight를 얻음.

feature engineering : 모델을 세우기에 앞서, 모델의 성능을 높일 수 있도록 feature들을 engineering함. one-hot encoding, class 나누기, 구간으로 나누기, 텍스트 데이터 처리 등을 함.

model 만들기 : sklearn을 사용해 모델을 만듦. 이를 이용하면 sklearn을 사용하면 수많은 알고리즘을 일관된 문법을 사용 가능함. 물론 딥러닝을 위해선 tensorflow, pytorch등을 사용 가능함.

모델 학습, 예측 : trainset을 가지고 모델을 학습시킨 후, testset을 가지고 prediction

모델 평가 : 예측 성능이 원하는 수준인지 판단. 풀려는 문제에 따라 모델을 평가하는 방식이 달라짐. 학습된 모델이 어떤 것을 학습했는지 확인.

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
#----데이터 분석에 있어서 기본적인 라이브러리 import----
import missingno as msno
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')
```

# 데이터셋 확인
##### Null data check

```python
for col in df_train.columns:
    #정형 데이터로 튜닝함
    msg = 'column: {:>10}\t Percent of NaN value: {:.2f}%'.format(col, 100*(df_train[col].isnull().sum()/df_train[col].shape[0]))
    print(msg)
```

```python
for col in df_test.columns:
    msg = 'column: {:>10}\t Percent of NaN value: {:.2f}%'.format(col, 100*(df_test[col].isnull().sum()/df_test[col].shape[0]))
    print(msg)
```

```python
msno.matrix(df=df_train.iloc[:,:], figsize=(8,8),color=(0.8,0.5,0.2))
```

```python
msno.bar(df=df_train.iloc[:,:], figsize=(8,8),color=(0.8,0.5,0.2))
```

##### Target label 확인 : 어떤 distribution을 가지고 있는지 확인해야 함

```python
f, ax = plt.subplots(1, 2, figsize=(18,8))

df_train['Survived'].value_counts().plot.pie(explode = [0,0.1], autopct='%1.1f%%', ax=ax[0], shadow=True)
ax[0].set_title('Pie chart - Survived')
ax[0].set_ylabel('')
#'Survived' 인자 값 입력할 때, 'x =' 붙이기 아니면 인자 충돌 일어나서 Type error남
sns.countplot(x='Survived', data=df_train, ax=ax[1])
ax[1].set_title('Bar chart - Survived')
```

# Exploratory data analysis
##### Pclass : original, 서수형 데이터. 카테고리이면서 , ordered data type

```python
#'Pclass'에 따른 생존률 차이. 엑셀의 피벗 차트와 유사한 작업을 하게 되는데, pandas dataframe에서는 groupby를 사용하면 쉽게 할 수 있음
#'pclass', 'survived'를 가져온 후, pclass로 묶음. 그 후, pclass마다 0,1이 count가 되는데, 이를 평균내면 각 pclass별 생존률이 나옴
df_train[['Pclass','Survived']].groupby(['Pclass'],as_index=True).count()
```

```python
#count()를 하면, 각 class에 몇 명이 있는지 확인할 수 있으며, sum()을 하면, 216명 중 생존한(survived=1) 사람의 총합을 구할 수 있음
df_train[['Pclass','Survived']].groupby(['Pclass'],as_index=True).sum()
```

```python
#pandas의 crosstab을 사용하면 위 과정을 좀 더 수월하게 볼 수 있음
pd.crosstab(df_train['Pclass'], df_train['Survived'],margins=True).style.background_gradient(cmap='summer_r')
```

```python
#grouped객체에 mean()을 하게 되면, 각 클래스별 생존률을 얻을 수 있음.
#Pclass가 좋을수록(1st) 생존률이 높음.
df_train[['Pclass','Survived']].groupby(['Pclass'],as_index=True).mean().sort_values(by='Survived',ascending=False).plot.bar()
```

```python
#클래스가 높을수록, 생존 확률이 높음
#생존에 Pclass가 큰 영향을 미친다는 걸 알 수 있음. 나중에 모델을 세울 때, 이 feature를 사용하는 것이 좋을 것이라 판단 가능.
y_position=1.02
f, ax= plt.subplots(1,2,figsize=(18,8))
df_train['Pclass'].value_counts().plot.bar(color=['#CD7F32','#FFDF00','#D3D3D3'],ax=ax[0])
ax[0].set_title('Number of Passengers By Pclass', y=y_position)
ax[0].set_ylabel('Count')
sns.countplot(x='Pclass',hue='Survived',data=df_train,ax=ax[1])
ax[1].set_title('Pclass: Survived vs Dead', y=y_position)
plt.show()
```

##### Sex : 성별별 생존률 확인

```python
#여자 생존확률 더 높음
f, ax = plt.subplots(1,2,figsize=(18,8))
df_train[['Sex','Survived']].groupby(['Sex'],as_index=True).mean().plot.bar(ax=ax[0])
ax[0].set_title('Survived as Sex')
sns.countplot(x='Sex',hue='Survived',data=df_train,ax=ax[1])
ax[1].set_title('Sex: Survived vs Dead')
plt.show()
```

### Age

```python
print('oldest : {:.1f}Years.'.format(df_train['Age'].max()))
print('Youngest : {:.1f}Years.'.format(df_train['Age'].min()))
print('Average : {:.1f}Years.'.format(df_train['Age'].mean()))
```

```python
fig, ax= plt.subplots(1,1,figsize=(9,5))
sns.kdeplot(df_train[df_train['Survived']==1]['Age'],ax=ax)
sns.kdeplot(df_train[df_train['Survived']==0]['Age'],ax=ax)
plt.legend(['Survived==1','Survived==0'])
plt.show()
```

```python
#Age distribution within classes
#class가 높을수록, 나이 많은 사람들의 비중이 커짐
plt.figure(figsize=(8,6))
df_train['Age'][df_train['Pclass']==1].plot(kind='kde')
df_train['Age'][df_train['Pclass']==2].plot(kind='kde')
df_train['Age'][df_train['Pclass']==3].plot(kind='kde')
plt.xlabel('Age')
plt.title('Age Distribution within classes')
plt.legend(['1st classes','2nd classes','3rd classes'])
```

```python
#나이가 어릴 수록, 생존률이 높음
cummulate_survival_ratio=[]
for i in range(1,80):
    cummulate_survival_ratio.append(df_train[df_train['Age']<i]['Survived'].sum()/len(df_train[df_train['Age']<i]['Survived']))
    
plt.figure(figsize=(7,7))
plt.plot(cummulate_survival_ratio)
plt.title('Survival rate change depending on range of Age',y=1.02)
plt.ylabel('Survival rate')
plt.xlabel('Range of Age(0~x)')
plt.show()
```

### Fare(탑승요금) 

```python
#skewness : 비대칭도
fig, ax =plt.subplots(1,1,figsize=(8,8))
g=sns.distplot(df_train['Fare'],color='b',label='Skewness: {:.2f}'.format(df_train['Fare'].skew()),ax=ax)
g=g.legend(loc='best')
```

```python
#testset에 있는 nan value를 평균값으로 치환함
#feature engineering에 들어가는 부분의 작업
df_test.loc[df_test.Fare.isnull(),'Fare']=df_test['Fare'].mean()
df_train['Fare']=df_train['Fare'].map(lambda i: np.log(i) if i>0 else 0)
df_test['Fare']=df_test['Fare'].map(lambda i: np.log(i) if i>0 else 0)
```

```python
fig,ax=plt.subplots(1,1,figsize=(8,8))
g=sns.distplot(df_train['Fare'],color='b',label='Skewness : {:.2f}'.format(df_train['Fare'].skew()),ax=ax)
g=g.legend(loc='best')
```

# feature engineering
##### Fill Null
title + statistics를 사용하여 채울 예정
1) data -> string으로 바꿔주는 str method
2) 정규표현식을 적용하게 해주는 extract method(추출)

```python
df_train['Initial'] =df_train.Name.str.extract('([A-Za-z]+)\.')
df_test['Initial'] =df_test.Name.str.extract('([A-Za-z]+)\.')
```

```python
#pandas의 crosstab을 이용하여 우리가 추출한 Initial과 Sex간의 count를 살펴볼 수 있음
pd.crosstab(df_train['Initial'],df_train['Sex']).T.style.background_gradient(cmap='summer_r')
```

```python
df_train['Initial'].replace(['Mlle','Mme','Ms','Dr','Major','Lady','Countess','Jonkheer','Col','Rev','Capt','Sir','Don','Dona'],['Miss','Miss','Miss','Mr','Mr','Mrs','Mrs','Other','Other','Other','Mr','Mr','Mr', 'Mr'],inplace=True)
```
